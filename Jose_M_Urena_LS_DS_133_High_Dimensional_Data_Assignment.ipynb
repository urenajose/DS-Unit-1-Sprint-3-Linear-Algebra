# Vector Transformations

## 4.1 Plug the corresponding unit vectors into each function. Use the output vectors to create a transformation matrix.

\begin{align}
p(\begin{bmatrix}x_1 \\ x_2 \end{bmatrix}) = \begin{bmatrix} x_1 + 3x_2 \\2 x_2 - x_1 \\  \end{bmatrix}
\\
\\
q(\begin{bmatrix}x_1 \\ x_2 \\ x_3\end{bmatrix}) = \begin{bmatrix} 4x_1 + x_2 + 2x_3 \\2 x_2 - x_1 + 3x_3 \\ 5x_1 - 2x_3 + x_2  \end{bmatrix}
\end{align}

p function

```
import numpy as np
```

```
def p(vect):
  "Function p"
  x1=vect[0][0]
  x2=vect[1][0]
  row1 = np.array([x1+3*x2])
  row2 = np.array([2*x2-x1])
  output_vector = np.array([row1, row2])
  return output_vector
```

```
vector1_x = np.array([[1],[0]]) #passing a vector at random in x position
Jx = p(vector1_x)
print(Jx)
```

```
vector1_y = np.array([[0],[1]]) #passing a vector at random in x position
Jy = p(vector1_y)
print(Jy)
```

```
J = np.column_stack([Jx,Jy])
print(J)
```

##q function

```
def q(vect):
  x1=vect[0][0]
  x2=vect[1][0]
  x3=vect[2][0]
  row1 = np.array([4*x1+x2+2*x3])
  row2 = np.array([2*x2-x1+3*x3])
  row3 = np.array([5*x1-2*x3+x2])
  output_vector = np.array([row1, row2,row3])
  return output_vector
```

```
vectorq_x = np.array([[1],[0],[0]])
Qx = q(vectorq_x)
print(Qx)
```

```
vectorq_y = np.array([[0],[1],[0]])
Qy = q(vectorq_y)
print(Qy)
```

```
vectorq_z = np.array([[0],[0],[1]])
Qz = q(vectorq_z)
print(Qz)
```

```
Q = np.column_stack([Qx,Qy,Qz])
print(Q)
print(pd.DataFrame(Q))
```

Not only does this dataset have decent amount columns to begin with (99), but in preparing the data for PCA we have also [one-hot-encoded](https://hackernoon.com/what-is-one-hot-encoding-why-and-when-do-you-have-to-use-it-e3c6186d008f#targetText=One%20hot%20encoding%20is%20a,the%20entry%20in%20the%20dataset.) all of the categorical variables. Don't worry too much about the mechanics of one-hot encoding right now, you will learn and experiment with a whole bunch of categorical encoding approaches in unit 2. 

The code below will read in the dataset and perform the one-hot encoding of the categorical variables. Start adding your PCA code at the bottom of the provided code.

```
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read Natinal Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)

print(national.shape)
national.head()
```

```
# Look at datatypes
# a lot of object datatypes even though they seem to be strings of numbers.
national.dtypes
```

```
# check for null values
national.isnull().sum().any()
```

```
# check for number of categorical vs numeric columns
cat_cols = national.columns[national.dtypes=='object']
num_cols = national.columns[national.dtypes!='object']

print(f'{len(cat_cols)} categorical columns')
print(f'{len(num_cols)} numerical columns')
```

```
# We're making a copy of our data in case we mess something up.
national_processed = national.copy()

# Categorically Encode our Variables:
# They need to all be numeric before we do PCA.
# https://pbpython.com/categorical-encoding.html

# Cast categorical columns to "category" data type
national_processed[cat_cols] = national_processed[cat_cols].astype('category')

national_processed.dtypes
```

```
# Replace all category cell values with their numeric category codes
for col in cat_cols:
  national_processed[col] = national_processed[col].cat.codes

print(national_processed.shape)
national_processed.head()
```

```
# Now we only ahve numeric columns (ints and floats)
national_processed.dtypes
```

```
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```
### Your Code Here
feature_cols=national_processed.columns[1:]
feature_cols[:5]
```

```
x = national_processed.loc[:, feature_cols].values
x = StandardScaler().fit_transform(x) 
```

```
print('{:.2f}'.format(np.mean(x)))
print(np.std(x))
```

```
normal_national_processed = pd.DataFrame(x,columns=feature_cols)
normal_national_processed.head(5)
```

```
pca_national_processed = PCA(n_components=2)
pca_data = pca_national_processed.fit_transform(x)
```

```
pca_nationalp_final = pd.DataFrame(data = pca_data,columns = ['PC1','PC2'])
print(pca_nationalp_final.head(5))
print(pca_nationalp_final.shape)
```
